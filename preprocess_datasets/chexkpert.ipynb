```python
CHEXPERT_TASKS = [
    "No Finding",
    "Enlarged Cardiomediastinum",
    "Cardiomegaly",
    "Lung Lesion",
    "Lung Opacity",
    "Edema",
    "Consolidation",
    "Pneumonia",
    "Atelectasis",
    "Pneumothorax",
    "Pleural Effusion",
    "Pleural Other",
    "Fracture",
    "Support Devices",
]
CHEXPERT_COMPETITION_TASKS = [
    "Atelectasis",
    "Cardiomegaly",
    "Consolidation",
    "Edema",
    "Pleural Effusion",
]

# baseed on original chexpert paper
CHEXPERT_UNCERTAIN_MAPPINGS = {
    "Atelectasis": 1,
    "Cardiomegaly": 0,
    "Consolidation": 0,
    "Edema": 1,
    "Pleural Effusion": 1,
}
```

```python
import numpy as np
CHEXPERT_ORIGINAL_TRAIN_CSV = '/u/home/lj0/datasets/cheXpert_small/CheXpert-v1.0-small/train.csv'
def preprocess_chexpert_5x200_data():
    df = pd.read_csv(CHEXPERT_ORIGINAL_TRAIN_CSV)
    df = df.fillna(0)
    df = df[df["Frontal/Lateral"] == "Frontal"]

    task_dfs = []
    for i in range(len(CHEXPERT_COMPETITION_TASKS)):
        index = np.zeros(14)
        index[i] = 1
        df_task = df[
            (df["Atelectasis"] == index[0])
            & (df["Cardiomegaly"] == index[1])
            & (df["Consolidation"] == index[2])
            & (df["Edema"] == index[3])
            & (df["Pleural Effusion"] == index[4])
            & (df["Enlarged Cardiomediastinum"] == index[5])
            & (df["Lung Lesion"] == index[7])
            & (df["Lung Opacity"] == index[8])
            & (df["Pneumonia"] == index[9])
            & (df["Pneumothorax"] == index[10])
            & (df["Pleural Other"] == index[11])
            & (df["Fracture"] == index[12])
            & (df["Support Devices"] == index[13])
        ]
        df_task = df_task.sample(n=200, random_state=0)
        task_dfs.append(df_task)
    df_200 = pd.concat(task_dfs)

    return df_200
```

```python
df_200 = preprocess_chexpert_5x200_data()
```

```python
CHEXPERT_PATH_COL = "Path"
df = df[~df[CHEXPERT_PATH_COL].isin(df_200[CHEXPERT_PATH_COL])]
valid_ids = np.random.choice(
    len(df), size=5000, replace=False) # ramdon select 5000 images
valid_df = df.iloc[valid_ids]
train_df = df.drop(valid_ids, errors="ignore")

print(f"Number of train samples: {len(train_df)}")
print(f"Number of valid samples: {len(valid_df)}")
print(f"Number of chexpert5x200 samples: {len(df_200)}")

train_df.to_csv('/u/home/lj0/datasets/cheXpert_small/CheXpert-v1.0-small/train_split.csv')
valid_df.to_csv('/u/home/lj0/datasets/cheXpert_small/CheXpert-v1.0-small/valid_split.csv')
df_200.to_csv('/u/home/lj0/datasets/cheXpert_small/CheXpert-v1.0-small/chexpert_5x200.csv')
```
